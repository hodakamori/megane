```python
import megane

pipe = megane.Pipeline()
s = pipe.add_node(megane.LoadStructure('tests/fixtures/caffeine_water.pdb'))
bonds = pipe.add_node(megane.AddBonds(source='distance'))
pipe.add_edge(s, bonds)

viewer = megane.MolecularViewer()
viewer.set_pipeline(pipe)
viewer
```

```python
import megane

pipe2 = megane.Pipeline()
s2 = pipe2.add_node(megane.LoadStructure('tests/fixtures/perovskite_srtio3.xyz'))
bonds2 = pipe2.add_node(megane.AddBonds(source='distance'))
poly = pipe2.add_node(megane.AddPolyhedra(
    center_elements=[22],   # Ti
    ligand_elements=[8],     # O
    max_distance=2.5,
    opacity=0.5,
    show_edges=True,
))
pipe2.add_edge(s2, bonds2)
pipe2.add_edge(s2, poly)

viewer2 = megane.MolecularViewer()
viewer2.set_pipeline(pipe2)
viewer2
```

```python
import megane

pipe3 = megane.Pipeline()
s3 = pipe3.add_node(megane.LoadStructure('tests/fixtures/caffeine_water.pdb'))
f = pipe3.add_node(megane.Filter(query="element == 'C'"))
m = pipe3.add_node(megane.Modify(scale=1.3, opacity=0.8))
labels = pipe3.add_node(megane.AddLabels(source='element'))
bonds3 = pipe3.add_node(megane.AddBonds(source='distance'))

pipe3.add_edge(s3, f)
pipe3.add_edge(f, m)
pipe3.add_edge(m, labels)
pipe3.add_edge(m, bonds3)

viewer3 = megane.MolecularViewer()
viewer3.set_pipeline(pipe3)
viewer3
```

```python
import time
time.sleep(5)  # Wait for widget to fully render

png_data = viewer.render_image(width=640, height=480)
print(f'PNG size: {len(png_data)} bytes')
assert len(png_data) > 1000, 'PNG data too small'
assert png_data[:8] == b'\x89PNG\r\n\x1a\n', 'Not a valid PNG'
print('Image render: OK')
```

```python
import megane

pipe4 = megane.Pipeline()
s4 = pipe4.add_node(megane.LoadStructure('tests/fixtures/caffeine_water.pdb'))
t = pipe4.add_node(megane.LoadTrajectory(xtc='tests/fixtures/caffeine_water_vibration.xtc'))
pipe4.add_edge(s4, t)

viewer4 = megane.MolecularViewer()
viewer4.set_pipeline(pipe4)
viewer4

# Note: render_video test is in a separate cell below to allow widget init
```

```python
import time
time.sleep(5)  # Wait for widget to fully render

gif_data = viewer4.render_video(width=320, height=240, fps=10, format='gif')
print(f'GIF size: {len(gif_data)} bytes')
assert len(gif_data) > 1000, 'GIF data too small'
assert gif_data[:6] in (b'GIF87a', b'GIF89a'), 'Not a valid GIF'
print('Video render: OK')
```
