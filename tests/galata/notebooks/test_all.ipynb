```python
import os
import pathlib

# Set CWD to project root so relative paths like 'tests/fixtures/...' work
_notebook_dir = pathlib.Path(os.getcwd())
_project_root = _notebook_dir
while _project_root != _project_root.parent:
    if (_project_root / 'pyproject.toml').exists():
        break
    _project_root = _project_root.parent
os.chdir(_project_root)

import megane

pipe = megane.Pipeline()
s = pipe.add_node(megane.LoadStructure('tests/fixtures/caffeine_water.pdb'))
bonds = pipe.add_node(megane.AddBonds(source='distance'))
pipe.add_edge(s, bonds)

viewer = megane.MolecularViewer()
viewer.set_pipeline(pipe)
viewer
```

```python
import megane

pipe2 = megane.Pipeline()
s2 = pipe2.add_node(megane.LoadStructure('tests/fixtures/perovskite_srtio3.xyz'))
bonds2 = pipe2.add_node(megane.AddBonds(source='distance'))
poly = pipe2.add_node(megane.AddPolyhedra(
    center_elements=[22],   # Ti
    ligand_elements=[8],     # O
    max_distance=2.5,
    opacity=0.5,
    show_edges=True,
))
pipe2.add_edge(s2, bonds2)
pipe2.add_edge(s2, poly)

viewer2 = megane.MolecularViewer()
viewer2.set_pipeline(pipe2)
viewer2
```

```python
import megane

pipe3 = megane.Pipeline()
s3 = pipe3.add_node(megane.LoadStructure('tests/fixtures/caffeine_water.pdb'))
f = pipe3.add_node(megane.Filter(query="element == 'C'"))
m = pipe3.add_node(megane.Modify(scale=1.3, opacity=0.8))
labels = pipe3.add_node(megane.AddLabels(source='element'))
bonds3 = pipe3.add_node(megane.AddBonds(source='distance'))

pipe3.add_edge(s3, f)
pipe3.add_edge(f, m)
pipe3.add_edge(m, labels)
pipe3.add_edge(m, bonds3)

viewer3 = megane.MolecularViewer()
viewer3.set_pipeline(pipe3)
viewer3
```
