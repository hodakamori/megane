```python
import megane

pipe = megane.Pipeline()
s = pipe.add_node(megane.LoadStructure('tests/fixtures/perovskite_srtio3.xyz'))
bonds = pipe.add_node(megane.AddBonds(source='distance'))
poly = pipe.add_node(megane.AddPolyhedra(
    center_elements=[22],   # Ti
    ligand_elements=[8],     # O
    max_distance=2.5,
    opacity=0.5,
    show_edges=True,
))
pipe.add_edge(s, bonds)
pipe.add_edge(s, poly)

viewer = megane.MolecularViewer()
viewer.set_pipeline(pipe)
viewer
```
