```python
import megane

pipe = megane.Pipeline()
s = pipe.add_node(megane.LoadStructure('tests/fixtures/caffeine_water.pdb'))
bonds = pipe.add_node(megane.AddBonds(source='distance'))
pipe.add_edge(s, bonds)

viewer = megane.MolecularViewer()
viewer.set_pipeline(pipe)
viewer
```

```python
import time
time.sleep(5)  # Wait for widget to fully render

png_data = viewer.render_image(width=640, height=480)
print(f'PNG size: {len(png_data)} bytes')
assert len(png_data) > 1000, 'PNG data too small'
assert png_data[:8] == b'\x89PNG\r\n\x1a\n', 'Not a valid PNG'
print('Image render: OK')
```
